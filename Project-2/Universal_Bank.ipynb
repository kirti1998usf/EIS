```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import cross_val_score
```

# Splitting and training the data

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Income','CD Account','CCAvg','Education','Mortgage']], df['Personal Loan'], test_size=0.3, random_state=100)
DTC=DecisionTreeClassifier()
```

# Kfold

```python
cross_val_score(estimator=DTC,X=X_train,y=y_train,cv=7)
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
params_dtc = [{'criterion': ['gini','entropy'],'max_depth':[2,3,4,5,6,7],'max_features':[1,2,3,4]}]
grid = GridSearchCV(estimator=DTC,param_grid=params_dtc,cv=10,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
DTC_best = DecisionTreeClassifier(criterion='entropy',max_depth=5,max_features=4)
```

# Model fitting 

```python
DTC_best.fit(X_train,y_train)
y_pred = DTC_best.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,confusion_matrix
```

```python
print("Accuracy score of model:",accuracy_score(y_test,y_pred))
```

```python
print("Precision score of model:",precision_score(y_test,y_pred))
```

```python
print("f1_score of model:",f1_score(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.show()
```
